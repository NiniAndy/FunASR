```python
import torch

root_path = "/ssd/zhuang/code/FunASR/examples/aishell/hypformer/exp/baseline_hypformer_conformer_12e_6d_2048_256_zh_char_exp1/model.pt.avg10"
root_model = torch.load(root_path, map_location='cpu')
root_model['state_dict'].keys()

# backup
torch.save(root_model, root_path + ".backup")
```

```python
# rename 
# encoder.embed.out.weight -> encoder.embed.out.0.weight
# encoder.embed.out.bias -> encoder.embed.out.0.bias

new_state_dict = {}
for k, v in root_model['state_dict'].items():
    if 'encoder.embed.out.weight' in k:
        k = k.replace('encoder.embed.out.weight', 'encoder.embed.out.0.weight')
    if 'encoder.embed.out.bias' in k:
        k = k.replace('encoder.embed.out.bias', 'encoder.embed.out.0.bias')
    new_state_dict[k] = v
    
root_model['state_dict'] = new_state_dict
torch.save(root_model, root_path)
```
