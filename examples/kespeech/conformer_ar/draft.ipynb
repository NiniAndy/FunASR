```python
# 为conformer_ar_only构建初始化参数
import torch

pre_trained_model = torch.load('/ssd/zhuang/code/FunASR/examples/kespeech/conformer/exp/baseline/model.pt.avg10')['state_dict']
pre_trained_model.keys()
```

```python
# 除了decoder.embed.0.weight其他的decoder参数都删掉 
# 并将decoder.embed.0.weight改名成ctc_tokenizer.0.weight
for key, value in list(pre_trained_model.items()):
    if key.startswith('decoder') and not key.startswith('decoder.embed.0.weight'):
        del pre_trained_model[key]
    if key.startswith('decoder.embed.0.weight'):
        pre_trained_model['ar_ctc_tokenizer.0.weight'] = value
        del pre_trained_model[key]
    if key.startswith('encoder.'):
        pre_trained_model[key.replace('encoder.', 'ar_encoder.')] = value
        del pre_trained_model[key]
    if key.startswith('ctc.'):
        pre_trained_model[key.replace('ctc.', 'ar_ctc.')] = value
        del pre_trained_model[key]
        
```

```python
# 为lora-ar-asr做初始化
import torch

# asr init model ckpt
asr_model = torch.load('/ssd/zhuang/code/FunASR/examples/kespeech/conformer/exp/baseline/model.pt.avg10')['state_dict']
asr_model.keys()

for key, value in list(asr_model.items()):
    if key.startswith('decoder'):
        del asr_model[key]
    if key.startswith('encoder.'):
        asr_model[key.replace('encoder.', 'asr_encoder.')] = value
        del asr_model[key]
    if key.startswith('ctc.'):
        asr_model[key.replace('ctc.', 'asr_ctc.')] = value
        del asr_model[key]
```

```python
# init ar model ckpt
ar_model = torch.load('/ssd/zhuang/code/FunASR/examples/kespeech/conformer_lora/exp/conformer_lora_WD2AR/model.pt.avg10')['state_dict']


for key, value in list(ar_model.items()):
    if key.startswith('ar_encoder.ar_encoder.'):
        ar_model[key.replace('ar_encoder.ar_encoder.', 'ar_decoder.ar_linear.')] = value
        del ar_model[key]
        
ar_model.keys()
```

```python
for key, value in list(ar_model.items()):
    if key.startswith('ar_encoder'):
        ar_model[key.replace('ar_encoder', 'ar_decoder')] = value
        ar_model[key] = value
        del ar_model[key]

ar_model.keys()
```

```python
for key, value in list(ar_model.items()):
    if key.startswith('share_encoder'):
        ar_model[key.replace('share_encoder', 'ar_encoder')] = value
        ar_model[key] = value
        del ar_model[key]

ar_model.keys()
```

```python
for key, value in list(ar_model.items()):
    if key.startswith('ar_ctc.'):
        ar_model[key.replace('ar_ctc.', 'ar_ctc.')] = value
        ar_model[key] = value
        del ar_model[key]
    if key.startswith('ar_ctc_tokenizer'):
        ar_model[key.replace('ar_ctc_tokenizer', 'ar_ctc_tokenizer')] = value
        ar_model[key] = value
        del ar_model[key]

ar_model.keys()
```

```python
torch.save(ar_model, '/ssd/zhuang/code/FunASR/examples/kespeech/conformer_lora/exp/conformer_lora_WD2AR/model.pt.avg10.rename')
```

```python
# 把asr和ar的参数合并
asr_model.update(ar_model)

torch.save(asr_model, '/ssd/zhuang/code/FunASR/examples/kespeech/conformer_lora/exp/conformer_lora_asrNar/init_model.pt')
```
